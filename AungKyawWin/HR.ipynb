```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\akw97\Downloads\PythonProgramming-AKW-\data\HR_5000 Records.csv")
df.info()
```

```python
df['Last % Hike'] = df['Last % Hike'].str.strip()
df['Last % Hike'] = df['Last % Hike'].str.replace('%','', regex = False)
df['Last % Hike'] = df['Last % Hike'].astype(float)
```

```python
df['Phone No. '] = df['Phone No. '].str.strip()
df['Phone No. '] = df['Phone No. '].str.replace('-','', regex = False)
df['Phone No. '] = df['Phone No. '].astype(float)
```

```python
df['SSN'] = df['SSN'].str.strip()
df['SSN'] = df['SSN'].str.replace('-','', regex = False)
df['SSN'] = df['SSN'].astype(float)
```

```python
df.describe()
```

```python
select_col  = ['Age in Company (Years)', 'Salary', 'Last % Hike', 'Age in Yrs.']
sub_df = df[select_col]
correlation = sub_df.corr()
print(correlation)

plt.figure(figsize = (10,8))
sns.heatmap(correlation, annot = True, cmap = 'coolwarm', fmt = ".2f", linewidths = .5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
gender_count = df['Gender'].value_counts()
print(gender_count)

plt.figure(figsize = (8,6))
sns.barplot(x=gender_count.index, y=gender_count.values, palette='pastel')

for index, value in enumerate(gender_count.values):
    plt.text(index, value + 0.1, str(value), ha='center', va='bottom')

# plt.title('Number of Employees Based on Gender')
# plt.xlabel('Gender')
# plt.ylabel('Number of Employees')
plt.show()
```

```python
region_count = df['Region'].value_counts()
print(region_count)

plt.figure(figsize=(8,8))
plt.pie(region_count, explode = (0.1,0,0,0), labels = region_count.index, autopct ='%1.1f%%', startangle = 90 )
# plt.title('Number of Employees Based on Region')
plt.show()
```

```python
gender_by_region = df.groupby('Region')['Gender'].value_counts()
descending_sorted = gender_by_region.sort_values(ascending = False)
print(descending_sorted)

plt.figure(figsize = (8,6))
descending_sorted.plot(kind = 'bar', color = ['skyblue', 'yellow'], alpha = 0.7)
# plt.title('Gender Distribution by Region')
# plt.xlabel = ('Region, Gender')
# plt.ylabel = ('Count')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

```python
gender_length_in_company = df.groupby('Gender')['Age in Company (Years)'].agg(['mean','min','max'])
print(gender_length_in_company)
plt.figure(figsize = (8,6))
gender_length_in_company.plot(kind = 'bar', colormap = 'viridis', alpha = 0.7)
# plt.title('Age in Company by Gender')
# plt.xlabel = ('Gender')
# plt.ylabel = ('Age in Company(Years)')
plt.show()
```

```python
gender_length_in_company = df.groupby('Gender')['Age in Company (Years)'].agg(['mean','min','max'])
print(gender_length_in_company)
plt.figure(figsize = (12,10))
df.boxplot(column = 'Age in Company (Years)', by = 'Gender', grid = False, patch_artist= True, medianprops = {'color':'black'})
plt.show()
```

```python
selected_columns = ['Name Prefix', 'Salary']

subset_df = df[selected_columns]

prefix_mean_salary = subset_df.groupby('Name Prefix')['Salary'].agg(['mean', 'min', 'max'])

sorted_prefix_mean_salary = prefix_mean_salary.sort_values(by =['mean'], ascending = False)

print(sorted_prefix_mean_salary)

sorted_prefix_mean_salary.plot(kind = 'bar', figsize = (12,6), colormap = 'viridis')

plt.xlabel = ('Name Prefix')
plt.title = ('Salary by Name Prefix')
plt.show()

plt.figure(figsize = (25,15))
df.boxplot(column = 'Salary', by = 'Name Prefix', grid = False, patch_artist = True, medianprops = {'color': 'black'})
plt.show()
```

```python
selected_columns = ['Age in Yrs.', 'Salary']

below_40_df = df[df['Age in Yrs.'] < 40][selected_columns]
above_40_df = df[df['Age in Yrs.'] > 40][selected_columns]


num_people_below_40 = below_40_df.shape[0]
num_people_above_40 = above_40_df.shape[0]

print(f"Number of people below 40: {num_people_below_40}")
print(f"Mean salary for people below 40: {below_40_df['Salary'].mean():.2f}\n")

print(f"Number of people above 40: {num_people_above_40}")
print(f"Mean salary for people above 40: {above_40_df['Salary'].mean():.2f}")
```
